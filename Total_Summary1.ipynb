<div class="viewer--28ZF"><h2>Когда могут потребоваться базы данных?</h2><hr><p class="p--1qUA">Для каждой компании важно принимать правильные бизнес-решения. Важно принимать их не на основе того, что "так чувствуется", а на основе данных.</p><p class="p--1qUA">В качестве примера возьмем интернет-магазин. В данном случае будут собираться данные:</p><ul><li class="p--1qUA">о визитах на сайт (напр. с помощью Яндекс.Метрики)</li><li class="p--1qUA">данные онлайн платежей (уже другая система)</li><li class="p--1qUA">данные по логистике (и еще одна)</li></ul><p class="p--1qUA">Как же здесь принимать решения? Пока используется всего один источник данных, вполне можно это делать в Excel, или в каких-то других системах аналитики. Но когда источников уже много, нужно как-то корректно совмещать из них данные, причём так, чтобы не было потерь, чтобы вся логика была правильной. С простой системой аналитики это сделать сложно. Тогда и начинаются SQL и базы данных!</p></div>

<div class="step__viewer--2j52"><div class="viewer--28ZF"><h2>Где хранят данные?</h2><p class="p--1qUA"><font color="#ff5533"><b>База данных</b></font> – это упорядоченный набор структурированной информации или данных, которые обычно хранятся в электронном виде в компьютерной системе.&nbsp;База данных обычно управляется системой управления базами данных (СУБД). Задача сохранять данные, логировать и складывать –&nbsp;причина появления СУБД.</p><p class="p--1qUA">Иными словами, в&nbsp;чем разница между базой данных (БД) и системой управления базами данных (СУБД)? База данных – условно, просто "файл", который хранит данные. СУБД – то, что записывает данные, читает, модифицирует, удаляет и т.д., через нее происходит работа с БД.</p><p class="p--1qUA">Данные вместе с СУБД, а также приложения, которые с ними связаны, называются системой баз данных, или, для краткости, просто базой данных.</p><p class="p--1qUA">СУБД могут быть <font color="#ff5533">реляционными</font> и <font color="#ff5533">NoSQL</font> (not only SQL).</p><hr><h3>Реляционные базы данных</h3><p class="p--1qUA">Элементы в реляционной базе данных организованы в виде набора таблиц со столбцами и строками. Технология реляционных баз данных обеспечивает наиболее эффективный и гибкий способ доступа к структурированной информации.</p><p class="p--1qUA">В реляционной базе данные разделены на таблицы, и одна таблица хранит однородные сущности, а связь между таблицами осуществляется некоторым общим ключом. Именно это свойство и отражено в названии, где слово реляционный происходит от англ. relation (отношения, зависимость).</p><p class="p--1qUA"><i>Пример:</i></p><div class="pb-2 pt-2 image-wrapper--3jTt"><div class=""><img loading="lazy" class="img-fluid" src="https://storage.yandexcloud.net/klms-public/production/learning-content/11/77/593/1702/7711/image.png"></div></div><hr><p class="p--1qUA">В целом, есть две версии происхождения названия:</p><ol><li class="p--1qUA">реляционные – значит, что в базе есть таблицы, между ними есть какие-то связи (relations)</li><li class="p--1qUA">реляционные – потому что к ним применяется реляционная алгебра (операции, направленные на работу с данными – объединить, пересечь, посчитать разность и т.д.)</li></ol><p class="p--1qUA">Еще одним свойством является поддержка транзакций.<span>&nbsp;</span><b>Транзакции</b><span>&nbsp;</span>– операции обработки данных, которые переводят базу из одного состояния в другое. Допустим, есть банковская система, нужно перевести деньги одного пользователя на счет другого, и вы последовательно выполняете операции – вычесть у одного, зачислить другому.&nbsp;</p><p class="p--1qUA">Транзакции подчиняются аббревиатуре&nbsp;<font color="#ff5533">ACID</font><span>&nbsp;</span>– атомарность, согласованность, изолированность и стойкость.&nbsp;</p><ul><li class="p--1qUA"><font color="#ff5533">Атомарность</font> – если мы что-то запустили (операции), то они либо все выполняются, либо все не выполняются.&nbsp;</li><li class="p--1qUA"><font color="#ff5533">Согласованность</font> – зафиксируются только допустимые результаты.</li><li class="p--1qUA"><font color="#ff5533">Изолированность</font> – если транзакции выполняются параллельно, то они не влияют друг на друга.</li><li class="p--1qUA"><font color="#ff5533">Стойкость</font> – если транзакция выполнена, то результат будет точно сохранен вне зависимости от проблем с оборудованием.</li></ul><hr><h3>NoSQL&nbsp;</h3><ol><li class="p--1qUA"><font color="#ff5533">Ключ-значение</font> – являются по сути словарем, позволяющим извлечь однозначное значение по ключу</li><li class="p--1qUA"><font color="#ff5533">Документо-ориентированные </font>– похожи на ключ-значение, только значения с разметкой (XML, JSON), которая и образует "документ"</li><li class="p--1qUA"><font color="#ff5533">Столбцовые</font> – данные хранятся по столбцам, а не по строкам</li><li class="p--1qUA"><font color="#ff5533">Графовые</font> – вершины, ребра и их свойства</li></ol><p class="p--1qUA">САР теорема (в распределенных системах) – NoSQL БД&nbsp;может обладать двумя свойствами из трех:</p><ul><li class="p--1qUA"><font color="#ff5533">Согласованность</font> (consistency)</li><li class="p--1qUA"><font color="#ff5533">Доступность</font> (availability)</li><li class="p--1qUA"><font color="#ff5533">Устойчивость к разделению</font> (partition tolerance)</li></ul></div></div>

<div class="viewer--28ZF"><h2>SQL –&nbsp;Structured Query Language</h2><p class="p--1qUA"><font color="#ff5533"><b>SQL</b></font>&nbsp;<font color="#ff5533">(Structured Query Language, структурированный язык запросов)</font> – это язык программирования, используемый в большинстве реляционных баз данных для запроса, обработки и определения данных, а также контроля доступа.&nbsp;</p><p class="p--1qUA">SQL является декларативным языком, то есть в нём не задается алгоритм решения задачи, а описывается, что нужно получить в качестве результата. В каком-то смысле SQL похож на обычную человеческую речь. Можно сказать, что это язык для заказа нужных данных и операций над ними.&nbsp;</p><p class="p--1qUA">Существует&nbsp;множество реализаций SQL, каждая из которых имеет свои&nbsp;особенности, но в общем схожа&nbsp;со стандартом SQL.&nbsp;В данном курсе будет рассмотрена система управления базами данных ClickHouse (и немного PostgreSQL).&nbsp;</p><hr><h3>ClickHouse</h3><p class="p--1qUA"><font color="#ff5533"><b>ClickHouse</b></font><span>&nbsp;</span>–&nbsp;столбцовая система управления базами данных (СУБД) для онлайн обработки аналитических запросов (OLAP) (<a href="https://clickhouse.tech/docs/ru/" target="_blank" class="view-links-processed">документация ClickHouse</a>). Главные особенности:</p><ul><li class="p--1qUA">Столбцовая база данных для аналитического использовния (OLAP)</li><li class="p--1qUA">Позволяет быстро читать большие данные</li><li class="p--1qUA">Транзакции отсутствуют</li><li class="p--1qUA">Данные добавляются, но не изменяются (изменения реализуется через инкремент)</li><li class="p--1qUA">Оконные функции появились совсем недавно (доверять им пока рано)</li></ul></div>

```python
t = """SELECT -- перечисление полей для результирующей таблицы
    BuyDate,
    COUNT(*) AS rows
FROM -- источник данных
    default.checks 
WHERE -- фильтрация данных
    BuyDate != '2019-03-08'
GROUP BY -- группировка данных
    BuyDate 
HAVING -- фильтрация данных, аналогично WHERE, при использовании GROUP BY
    COUNT(*) > 215000
ORDER BY -- сортировка результирующей таблицы
    rows DESC 
LIMIT 100 -- ограничение на кол-во строк результирующей таблицы """
```

<div class="lesson--24hc"><div class="wrapper--3aUs"><span><a aria-current="page" class="link--1XS6 active" href="/learning/320"> Курс «Аналитика Данных B2B» /</a> </span><span><a aria-current="page" class="link--1XS6 active" href="/learning/320/module/3128"> Модуль «SQL» /</a> </span><span class="text--3Rvs"> Урок «Группировка и агрегатные функции» Конспект </span></div><div class="row h100--7AQS"><div class="col col-12 h100--7AQS"><div class="step__grid--3NLC"><div class="step__viewer--2j52"><div class="viewer--28ZF"><h2>Агрегатные функции</h2><p class="p--1qUA"><font color="#ff5533"><b>Агрегатные функции</b></font><span>&nbsp;</span>– функции, которые помогают как-либо обобщить данные.&nbsp;<br></p><p class="p--1qUA">Агрегатные функции используются в блоке<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">SELECT</code></code><span>&nbsp;</span>и в качестве<span>&nbsp;</span><b>параметра</b><span>&nbsp;</span>принимают наименования поля (в примере –&nbsp;<code><code class="klms-inline-code" data-lang="common">column</code></code>). К ним относятся:</p><ul><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">COUNT()</code></code><span>&nbsp;</span>–&nbsp;количество значений поля<span>&nbsp;</span>(без параметра считает кол-во строк, с полем – кол-во без NULL значений)</li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">COUNT(DISTINCT column)</code></code><span>&nbsp;</span>–&nbsp;количество уникальных значений поля (<code><code class="klms-inline-code" data-lang="common">uniqExact()</code></code><span>&nbsp;</span>и<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">count(DISTINCT)</code></code>&nbsp;– одно и тоже​​​​​)</li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">AVG(column)</code></code><span>&nbsp;</span>– вычисляет среднее<br></li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">MIN(column)</code></code><span>&nbsp;</span>–&nbsp;показывает минимальное значение</li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">MAX(column)</code></code><span>&nbsp;</span>–&nbsp;показывает максимальное значение</li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">SUM(column)</code></code><span>&nbsp;</span>–&nbsp;показывает сумму элементов</li></ul><p class="p--1qUA">Такие функции вычисляют заказанное значение на уровне группировки, заданном в секции<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">GROUP BY</code></code>.<br><br></p><h2>Примеры работы</h2><hr><p class="p--1qUA">1. <div class="row h100--7AQS"><div class="col col-12 h100--7AQS"><div class="step__grid--3NLC"><div class="step__viewer--2j52"><div class="viewer--28ZF"><h2>Синтаксис оконных функций (продолжение)</h2><p class="p--1qUA">При создании окна ему можно передать следующие параметры:</p><ul><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">PARTITION BY</code> <code class="klms-inline-code" data-lang="common">[city]</code></code></li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">ORDER BY</code> <code class="klms-inline-code" data-lang="common">[date ASC]</code></code></li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">ROWS [BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW]</code></code></li></ul><p class="p--1qUA">–&nbsp;<code class="klms-inline-code" data-lang="common">UNBOUNDED PRECEDING</code>&nbsp;– указывает, что окно начинается с первой строки секции</p><p class="p--1qUA">–&nbsp;<code class="klms-inline-code" data-lang="common">UNBOUNDED FOLLOWING</code>&nbsp;– указывает, что окно заканчивается на последней строке секции</p><p class="p--1qUA">Давайте рассмотрим эти параметры подробнее!</p><hr><h2>PARTITION BY</h2><p class="p--1qUA">Группирует<span>&nbsp;</span>строки запроса в разделы, которые потом обрабатываются оконной функцией независимо друг от друга. Работает аналогично блоку<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">GROUP BY</code></code><span>&nbsp;</span>в запросе, только группирует данные для окна.</p><hr><h2>ORDER BY</h2><p class="p--1qUA">Сортирует<span>&nbsp;</span>результаты внутри раздела, тем самым определяя порядок, в котором оконная функция будет работать со строками. Работает аналогично блоку<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">ORDER BY</code></code><span>&nbsp;</span>на уровне запроса.</p><hr><h2>ROWS/RANGE</h2><p class="p--1qUA">Задает<span>&nbsp;</span>параметры<span>&nbsp;</span>рамки<span>&nbsp;</span>окна. Рамки используются в тех оконных функциях, которые работают с рамками, а не с разделом целиком. Первым аргументом задается начало рамки, вторым аргументом конец рамки, и дополнительно, третьим аргументом задается исключение из рамки.</p><p class="p--1qUA">Рамку можно задать в нескольких&nbsp;режимах, а именно:</p><ul><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">ROWS</code></code><span>&nbsp;</span>– Начало и конец рамки задаются в строках относительно текущей строки. Например<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">ROWS BETWEEN 3 PRECEDING AND 3 FOLLOWING</code></code><span>&nbsp;</span>означает создание рамки на 3 строки вверх и вниз относительно текущей строки.</li><li class="p--1qUA"><code><code class="klms-inline-code" data-lang="common">RANGE</code></code><span>&nbsp;</span>– Начало и конец рамки задаются в разнице значений столбца из<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">ORDER BY</code></code>. Например, если в<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">ORDER BY</code></code><span>&nbsp;</span>находится столбец<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">event_date</code></code><span>&nbsp;</span>с типом данных date, то определение окна можно задать следующим образом<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">RANGE BETWEEN '3 day' PRECEDING AND '3 days' FOLLOWING</code></code><span>&nbsp;</span>что будет означать рамку на 3 дня назад и вперед.</li></ul><p class="p--1qUA">При указании рамки через<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">RANGE</code></code><span>&nbsp;</span>обязательным условием является только один столбец в<span>&nbsp;</span><code><code class="klms-inline-code" data-lang="common">ORDER BY</code></code><span>&nbsp;</span>окна.</p></div></div></div><div class="footer--1BMP"><div class="footer_content--wj_F"><div class="nav-buttons--MGsQ"><button class="navigator__button--1imB">назад</button><button class="navigator__button--1imB">далее</button></div></div></div></div></div>

Чтобы не усложнять, остановимся на относительно простом понимании термина через такой формализм:&nbsp;</p><p class="p--1qUA"><code class="view-katex-processed" data-lang="latex"><span class="katex"><span class="katex-html" aria-hidden="true"><span class="base"><span class="strut" style="height: 0.625em; <div class="row h100--7AQS"><div class="col col-12 h100--7AQS"><div class="step__grid--3NLC"><div class="step__viewer--2j52"><div class="viewer--28ZF"><h2>Генеральная совокупность</h2><p class="p--1qUA"><font color="#ff5533"><b>Генеральная совокупность</b></font>&nbsp;–&nbsp;совокупность всех объектов (единиц), относительно которых предполагается делать выводы при изучении конкретной задачи.</p><p class="p--1qUA"><i>Пример: уровень тестостерона всех мужчин, участвующих в Олимпийских играх</i></p><p class="p--1qUA">Зачастую очень сложно&nbsp;исследовать все объекты, поэтому из генеральной совокупности берут <font color="#ff5533">выборки</font>.</p><p class="p--1qUA">Важной характеристикой выборки является её <font color="#ff5533">репрезентативность</font>. Под этим термином понимается соответствие характеристик выборки характеристикам генеральной совокупности в целом.</p><h2>Виды выборок</h2><p class="p--1qUA"><font color="#ff5533"><b>1. Вероятностные выборки</b></font>&nbsp;— при создании таких 
выборок мы предполагаем, что генеральная совокупность&nbsp; однородна и все её элементы одинаково доступны.</p><p class="p--1qUA"><i><font color="#ff5533">Простая случайная выборка</font></i> (simple random sample)&nbsp;–&nbsp;случайный набор объектов из генеральной совокупности.
Пример:&nbsp;100 мужчин, участвующих в Олимпийских играх</p><p class="p--1qUA"><i><font color="#ff5533">Стратифицированная выборка</font></i> (stratified 
sample)&nbsp;–&nbsp;перед тем, как случайным образом отобрать объекты из 
генеральной совокупности, мы разбиваем её&nbsp;на несколько страт (групп).
Пример: мужчины 18-25 лет, 26-31, 32-36 и так далее. Потом уже из этих групп случайно набираем по N человек.</p><p class="p--1qUA"><i><font color="#ff5533">Групповая выборка (cluster sample)</font></i>&nbsp;– сначала делим генеральную совокупность на кластеры, только считаем, что они между собой схожи.
Пример: рост жителей Санкт-Петербурга. Мы делим их на районы (Адмиралтейский, Василеостровский и т.д), а потом случайно набираем людей из нескольких случайно выбранных районов для исследования.</p><p class="p--1qUA"><font color="#ff5533"><b>2. Невероятностные выборки</b></font>&nbsp;— отбор в такой выборке 
осуществляется не по принципам случайности, а по субъективным критериям –
 доступности объектов, типичности или равного представительства. Такие 
выборки часто встречаются в социологических исследованиях, однако полученные на них данные обладают меньшей достоверностью и лучше их обходить 
стороной.
</p><p class="p--1qUA"><i><font color="#ff5533">Метод снежного кома</font></i>&nbsp;— у каждого респондента, начиная с 
первого, просятся контакты его друзей, коллег, знакомых, которые 
подходили бы под условия отбора и могли бы принять участие в 
исследовании.&nbsp; Основная проблема такой выборки — затрагивается не 
случайная группа лиц, а лица, связанные общими интересами, хобби и т.д.
</p><p class="p--1qUA"><font color="#ff5533"><i>Стихийная выборка</i></font>&nbsp;— производится опрос наиболее 
доступных респондентов. Размер и состав стихийных выборок заранее не 
известен и определяется только одним параметром - активностью 
респондентов.
Пример: опрос, проведенный в газете или журнале, большинство интернет-опросов
</p><p class="p--1qUA"><i><font color="#ff5533">Выборка типичных случаев</font></i>&nbsp;— происходит отбор отдельных единиц генеральной совокупности, которые 
обладают типичным значением признака (часто это среднее значение). При 
этом возникает проблема выбора признака и определения его типичного 
значения.</p></div></div></div><div class="footer--1BMP"><div class="footer_content--wj_F"><div class="nav-buttons--MGsQ"><button class="navigator__button--1imB">назад</button><button class="navigator__button--1imB">далее</button></div></div></div></div></div>

 и так далее. Мы не знаем, насколько различаются результаты участников

 из таких примеров является переменная возраста.
</p><p class="p--1qUA">Количественная непрерывная - возраст, измеренный в днях/месяцах/годах
</p><p class="p--1qUA">Ранговая переменная - возраст разбит на группы (очень часто встречается в анкетировании) - от 14-17 лет; 18-29 лет и так далее.</p></div></div></div><div class="footer--1BMP"><div class="footer_content--wj_F"><div class="nav-buttons--MGsQ"><button class="navigator__button--1imB">назад</button><button class="navigator__button--1imB">далее</button></div></div></div></div></div></div></div></div>

Если мы будем многократно извлекать выборки по N наблюдений из генеральной совокупности и внутри каждой выборки рассчитывать среднее значение и стандартное&nbsp;отклонение, то заметим, что распределение признака будет изменяться от выборки к выборке, при этом значения средних также будет варьироваться в положительную или отрицательную сторону.</p><p class="p--1qUA">Далее мы строим распределение выборочных средних значений. Если в каждой выборке оценка среднего не является точной, то как раз среднее всех средних будет очень близко к реальному среднему в генеральной совокупности. Большинство всех средних будет лежать рядом с нулем, а какие-то – отклоняться.&nbsp;</p><div><p style="text-align: center;"><img alt="" height="297" src="https://ucarecdn.com/9d276f73-2653-4e22-90eb-3691c02bccc0/" width="485"></p></div><p class="p--1qUA">Стандартное отклонение этого распределения называется <font color="#ff5533">стандартной ошибкой среднего</font>. Однако при достаточных размерах выборок и повторных их извлечений выборочное распределение средних всё равно будет нормальным! Вместо того, чтобы полагаться на какие-то готовые теоретические распределения показателей и считать статистическую значимость на их основе, бутстрап полагается исключительно на сами данные.</p><p class="p--1qUA">Как именно это работает?
